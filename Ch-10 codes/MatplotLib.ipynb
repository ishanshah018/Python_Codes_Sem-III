```python
# Draw a line in diagram from position (1,3) to (8,10)
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([1,8])
y1=np.array([3,10])
plt.plot(x1,y1)
plt.show()
```

```python
# Draw two points in diagram, one at (10,30) & one at (5,20)
x=np.array([5,10])
y=np.array([20,30])
plt.plot(x,y,'o')
plt.show()
```

```python
# Add Lables to x,y axis and also title

x=np.array([10,50])
y=np.array([15,45])
plt.plot(x,y)

plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")
plt.title("Sports Watch Data")
plt.show()
```

```python
# Use of Font Properties

x=np.array([80,150])
y=np.array([100,230])
plt.plot(x,y)

f1={'family':'serif', 'color':'blue','size':20 }
f2={'family':'serif', 'color':'darkred','size':15 }

plt.title("Spors watch Data",fontdict=f1)
plt.xlabel("Average Pulse",fontdict=f2)
plt.ylabel("Calories Burnt",fontdict=f2)
```

```python
# Use of Line Styles

x=np.array([80,150])
y=np.array([100,230])
plt.plot(x,y,color="red",
             linestyle=":", linewidth=3)

plt.title("Spors watch Data",fontweight="bold")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnt")
```

```python
# Draw two lines by specifying a plt.plot() for each line
y1=np.array([3,8,1,10])  # y values of line 1
y2=np.array([6,2,7,11])  # y values of line 2
# Here x plotting will be done by default values
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
# Draw two lines by specifying a plt.plot() for each line

x1=np.array([0,6])
y1=np.array([0,7])  
y2=np.array([0,2]) 
plt.plot(x1,y1,linestyle=":",linewidth=2)
plt.plot(x1,y2,linestyle=":",linewidth=2)
plt.show()
```

```python
# Markers

x=np.array([0,2,3,5])
y=np.array([2,7,6,8])
plt.plot(x,y,marker='o',ms=10)
```

```python
# fmt Format (Shortcut)
#  marker|line|color

x=np.array([0,2,3,5])
y=np.array([2,7,6,8])
plt.plot(x,y,'o:r')
plt.show()
```

```python
# Subplot (Column wise)
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
# Subplot (Row wise) On top of each other
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,1,1)
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)
```

```python
# Draw 6 Plots

x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(3,2,1)
plt.plot(x,y1)

plt.subplot(3,2,2)
plt.plot(x,y2)

plt.subplot(3,2,3)
plt.plot(x,y2)

plt.subplot(3,2,4)
plt.plot(x,y2)

plt.subplot(3,2,5)
plt.plot(x,y2)

plt.subplot(3,2,6)
plt.plot(x,y2)
```

```python
# Bar Plot

student={"Ishan":87,"Jack":55,"Sam":44}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```

```python
# Bar Plot

student={"Ishan":87,"Jack":55,"Sam":44}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar Plot

student={"Ishan":87,"Jack":55,"Sam":44}
names=list(student.keys())
values=list(student.values())
plt.barh(names,values,color="grey")
plt.show()
```

```python
# Scatter plot
x=[10,20,30,40,50,60,70]
y=[0,4,7,8,4,12,15]
plt.scatter(x,y)
plt.show()
```

```python
# Scatter plot with marker
x=[10,20,30,40,50,60,70]
y=[0,4,7,8,4,12,15]
a=[2,3,4,7,3,6,10]
plt.scatter(x,y,marker="*",s=200)
plt.scatter(x,a,marker=".",s=500)
plt.show()
```

```python
# Histogram

data=[1,2,3,3,3,3,9,9,5,4,4,8,8,8,6,7]
plt.hist(data)
plt.show()
```

```python
# PieChart
y=np.array([35,25,15,25])
mylabels=["Apples","Bananas","Cherries","Papaya"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
# PieChart
y=np.array([35,25,15,25])
mylabels=["Apples","Bananas","Cherries","Papaya"]
plt.pie(y,labels=mylabels,autopct="%0.1f%%")
plt.show()
```
